```python
# I am creating a list that contains all of the categories we will be using in the pie chart
food_cat_lis = list(food_categories_df.columns.values)
del food_cat_lis[0]
food_cat_lis
```

```python
# Going create a function to create the Pie Graphs
def pie_graph(sizes, categories, title):    

    plt.pie(sizes, shadow=True, startangle=180) 
    labels = [f'{l}, {s:0.1f}%' for l, s in zip(categories, sizes)]
    plt.title(title) 
    plt.tight_layout()
    plt.legend(title= 'Food Categories', labels = labels, loc='best', bbox_to_anchor=(-0.1, 1.))
    plt.show()
```

```python
sizes = [25,25,25,25]
categories = ['red', 'blue', 'white', 'green']
title= "It works"

pie_graph(sizes, categories, title)
```
